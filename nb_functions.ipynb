## Paramètres en nombre variable : `*`ou `**`
- `*iterables` : autant de paramètres que l'on veut par position 
    * le dernier par position
    * nom et type connu (bonne sémantique)
    * traitement homogène
- `**kwargs` : autant de paramètres que l'on veut par keyword 
    * dernier de la fonction
    * aucune idée du nom et type de paramètre (cf docstring, example)

```python
print()                  # 0 param par position 
print("un")              # 1 param par position
print("un", 2)           # 2 params par position
print("un", 2, [], {})   # 3 params par position

print("un", 2, [], {}, sep=' ; ', end='\n\n\n') # 3 params par position + 2 par keyword
```

```python
data = ["Toulouse", 123, True, (12.5, 34.6)]
print(*data, sep='\n', end='\n-----------\n')

# NB: appelle print en fournissant toutes les valeurs de la liste 
# comme paramètre par position
```

```python
alphabet = "abcdefghijklmnopqrstuvwxyz"
r = range(100)
cities = ['Toulouse', 'Pau', 'Saint-Brieuc', 'Lens', 'Aix-en-Provence', 'Rambouillet', 'Paris']

for i, l, city in zip(r, alphabet, cities):
    print(f"- {i}{l} - {city}")
```

```python
for x in zip(*cities):
    print(x)
```

```python
from itertools import zip_longest


for x in zip_longest(*cities, fillvalue='.'):
    print(x)
```

### Debug function with stars

```python
def f(a, b, *values, d=True, e=None, **kwargs):
    print(f"a = {a}")
    print(f"b = {b}")
    print(f"values = {values}") # tuple
    for v in values:
        print("- handle positional arg:", v)
    print(f"d = {d}")
    print(f"e = {e}")
    print(f"kwargs = {kwargs}")  # dict
    for kwarg_name, kwarg_value in kwargs.items():
        # print("- handle optional/keyword arg:", f"{kwarg_name} = {kwarg_value}")
        match kwarg_name:
            case 'color':
                print("- handle optional/keyword color:", kwarg_value)
            case 'width':
                if not isinstance(kwarg_value, int):
                    raise ValueError("width must be an integer")
                print("- handle optional/keyword width:", kwarg_value)
            case _:
                print("[warning] unknown argument:", kwarg_name)
                # raise ValueError(f"unknown argument: {kwarg_name}")
    print()
```

```python
f(1, 2)
```

```python
f(1, 2, 3, 4, 5, 6, color='red', width=4, d=False, e=[12, 15], meteo="très chaud")
```

```python
f(1, 2, 3, 4, 5, 6, color='red', width="thin", d=False, e=[12, 15], meteo="très chaud")
```
