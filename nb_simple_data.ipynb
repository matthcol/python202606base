```python
for expression in 'ou', 'lou', 'pou':
    print('Début recherche')
    if expression in city:
        position = city.index(expression)
        print(expression, 'trouvée à la position', position)
    else:
        print(expression, 'non trouvée')
    print('Fin recherche')
    print()
```

```python
for expression in 'ou', 'lou', 'pou', 'se':
    print('Début recherche')
    try:
        position = city.index(expression)
        print(expression, 'trouvée à la position', position)
    except ValueError:
        print(expression, 'non trouvée')
    print('Fin recherche')
    print()
print('The end')
```

```python
for city in 'Toulouse', "L'Haÿ-les-Roses", '東京':
    print('Ville :', city)
    print('Majuscule :', city.upper())
    for encoding in 'utf-8', 'cp1252', 'ISO-8859-15', 'ISO-8859-1', 'ISO-8859-6':
        try:
            code = city.encode(encoding)
            print('Coding : ', encoding, code)
        except Exception:
            print('non encodable en', encoding)
    print()
```

```python
for v in villes:
    print(v, end=' : ')
    if len(v) < 5:
        print('catégorie M5')
    elif len(v) < 10:
        print('catégorie M5-10')
    else:
        print('catégorie M10+')
```

```python
# pattern matching (match ... case) pour Python 3.10+
villes_123 = []
villes_456 = []
villes_789 = []
villes_10plus = []
for v in villes:
    match len(v):
        case 1 | 2 | 3:
            villes_123.append(v)
        case 4 | 5 | 6:
            villes_456.append(v)
        case 7 | 8 | 9:
            villes_789.append(v)
        case _:
            villes_10plus.append(v)
print('Catégorie 1-3 :', villes_123)
print('Catégorie 4-6 :', villes_456)
print('Catégorie 7-9 :', villes_789)
print('Catégorie 10+ :', villes_10plus)
```
