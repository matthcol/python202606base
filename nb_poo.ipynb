## Données temporelles
- module datetime, ISO 8601
- types : date, time, datetime


```python
# constructeur = fonction qui a le même nom que la classe
d = date(2028, 2, 29)
t = time(9, 45)
dt = datetime(2000, 2, 29, 12, 30, 15, 123456)
print(d)
print(t)
print(dt)
```

```python
for o in d, t, dt:
    print(o, 'has type', type(o))
```

```python
# tableau des formats personnalisés de date
# https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior
print(d.strftime('%d/%m/%Y'))
print(d.strftime('%d/%m/%y'))

print(f"date : {d:%d/%m/%Y}")  # calls .strftime(format)
print(f"date heure : {dt:%H:%M:%S %d/%m/%Y}")
```

```python
tz_bra = ZoneInfo("America/Sao_Paulo")

# méthode de classe
d2 = date.today()
dt2 = datetime.now()
dt3 = datetime.now(tz=tz_bra)

texte = "10:15 17/06/2026"
dt4 = datetime.strptime(texte, "%H:%M %d/%m/%Y")

# méthode objet
t2 = dt2.time()

print(d2)
print(dt2)
print(dt3)
print(dt4)
print(t2)
```

## Movies
classe métier

```python
from movie import Movie


m = Movie("Interstellar", 2014)
print(type(m))
print(m) # str
m # repr
```

```python
m2 = Movie("Mon Oncle", 1958, 116) # paramètres par position (positional)
m3 = Movie(title="Dune: Part One", year=2021, duration=155) # paramètres par nom (keyword)
m4 = Movie("Princess Bride", 1987, duration=98)
m5 = Movie("The Mandalorian and Grogu", 2026)
```

```python
movies = [m, m2, m3, m4, m5]
for movie in movies:
    print(movie) # str
    print(repr(movie))
    print("\t- title:", movie.title)
    print("\t- year:", movie.year)
    print("\t- duration:", movie.duration if movie.duration is not None else 'unknown')
```
