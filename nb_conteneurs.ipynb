```python
import math


def cartesien_to_polaire(coords_cartesiennes: tuple[float, float]) -> tuple[float, float]:
    x, y = coords_cartesiennes # unpack
    module = math.hypot(x, y)
    argument = math.atan2(y, x)
    return (module, argument)
```

```python
coords = (3.0, 4.0)
m, a = cartesien_to_polaire(coords)
print(f"module = {m} ; argument = {a}")
```
