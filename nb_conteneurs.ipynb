```python
import math


def cartesien_to_polaire(coords_cartesiennes: tuple[float, float]) -> tuple[float, float]:
    x, y = coords_cartesiennes # unpack
    module = math.hypot(x, y)
    argument = math.atan2(y, x)
    return (module, argument)
```

```python
coords = (3.0, 4.0)
m, a = cartesien_to_polaire(coords)
print(f"module = {m} ; argument = {a}")
```

### Exercice list + tuple
Liste de villes en tant que tuple (nom, code postal, population)

```python
villes
```

```python
villes_t: list[tuple[str, str, int]] = [
    ("Marseille", "13000", 886040),
    ("Nice", "06000", 357737),
    ("Toulon", "83000", 179116),
    ("Aix-en-Provence", "13100", 149695),
    ("Avignon", "84000", 92188),
    ("Antibes", "06600", 77637),
    ("Cannes", "06400", 74350),
    ("La Seyne-sur-Mer", "83500", 63732),
    ("Fréjus", "83600", 59719),
    ("Hyères", "83400", 55858),
]
```

- faire un pretty print de la liste de ville
- extraire les villes de plus de 100_000 habitants
- idem avec les noms uniquement dans le résultat
- population totale

```python
for ville_t in villes_t:
    print(ville_t)
```

```python
print("Villes :")
for nom, code_postal, population in villes_t:
    print(f" - {nom}, code postal {code_postal}, population {population}")
```

```python
for nom, code_postal, population in villes_t:
    if population > 100000:
        print(f"- {nom}, code postal {code_postal}, population {population}")
```

```python
# extraction
liste100k = []
for ville_t in villes_t:
    nom, code_postal, habitants = ville_t
    if habitants >= 100000 :
        liste100k.append(ville_t)

# display
print("liste des villes de plus de 100 000 habitants :")
for nom, code_postal, habitants in liste100k:
    print(f"{nom} ({code_postal}), {habitants} habitants")
```

```python
# extraction
liste100k = [
    (nom, code_postal, habitants) 
    for nom, code_postal, habitants in villes_t
    if habitants >= 100_000
]

# display
print("liste des villes de plus de 100 000 habitants :")
for nom, code_postal, habitants in liste100k:
    print(f"{nom} ({code_postal}), {habitants} habitants")
```
