## Listes

```python
# liste homogène (cas le plus usuel)
villes = ['Toulouse', 'Pau', 'Saint-Brieuc', 'Lens', 'Aix-en-Provence', 'Rambouillet', 'Paris']
villes
```

```python
# NB: annotation type => villes: list[str]
type(villes)
```

```python
# techniquement on peut faire une liste hétérogène
objets = [12, 3.14, "Toulouse", True, None]
objets
```

### Modification d'une liste

```python
villes.append('Marseille')
villes
```

```python
villes.extend(["Nantes", "Bordeaux", "Bayonne"])
villes
```

```python
villes.insert(0, 'Montpellier')
villes
```

```python
villes.remove('Pau')
villes
```

```python
# villes.clear()
# villes # => []
```

```python
villes
```

### Accès avec l'operateur []
En mode lecture, affectation et suppresion 

```python
print(villes)

# 1 element
print(villes[0])
print(villes[-1])

# 1 slice
print(villes[:3])
print(villes[3:6])
print(villes[6:])
print(villes[-3:])
print(villes[::2])
```

```python
villes[0] = 'Strasbourg'
villes
```

```python
# remplacer les 3 dernières par 2 autres
villes[-3:] = ['Lyon', 'Brest']
villes
```

```python
del villes[6]
villes
```

```python
del villes[-3:]
villes
```

### Itération

```python
for ville in villes:
    print(ville)
```
