## Listes

```python
# liste homogène (cas le plus usuel)
villes = ['Toulouse', 'Pau', 'Saint-Brieuc', 'Lens', 'Aix-en-Provence', 'Rambouillet', 'Paris']
villes
```

```python
# NB: annotation type => villes: list[str]
type(villes)
```

```python
# techniquement on peut faire une liste hétérogène
objets = [12, 3.14, "Toulouse", True, None]
objets
```

### Modification d'une liste

```python
villes.append('Marseille')
villes
```

```python
villes.extend(["Nantes", "Bordeaux", "Bayonne"])
villes
```

```python
villes.insert(0, 'Montpellier')
villes
```

```python
villes.remove('Pau')
villes
```

```python
villes
```

### Accès avec l'operateur []
En mode lecture, affectation et suppresion 

```python
print(villes)

# 1 element
print(villes[0])
print(villes[-1])

# 1 slice
print(villes[:3])
print(villes[3:6])
print(villes[6:])
print(villes[-3:])
print(villes[::2])
```

```python
villes[0] = 'Strasbourg'
villes
```

```python
# remplacer les 3 dernières par 2 autres
villes[-3:] = ['Lyon', 'Brest']
villes
```

```python
del villes[6]
villes
```

```python
del villes[-3:]
villes
```

### Itération

```python
for ville in villes:
    print(ville)
```

```python
# calculer le nombre total de lettres de toutes les villes
total_letters = 0
for ville in villes:
    total_letters += len(ville)   # total_letters = total_letters + len(ville)
total_letters
```

```python
print('Nb de lettres (str) :', len(ville))
print('Nb de villes (list) :', len(villes))
```

```python
from collections.abc import Iterable, Sized


assert type(ville) is str
assert type(villes) is list

# str est à la fois Sized (len) est Iterable (for, ...) 
assert isinstance(ville, Sized)
assert isinstance(ville, Iterable)

# list est à la fois Sized (len) est Iterable (for, ...)
assert isinstance(villes, Sized)
assert isinstance(villes, Iterable)
```

```python
nombre = 123
assert not isinstance(nombre, Sized)
assert not isinstance(nombre, Iterable)

# len(nombre) # erreur de typage: nombre n'est pas Sized
# for x in nombre:  # erreur de typage : nombre n'est pas Iterable
#     print(x)
```

```python
for i, ville in enumerate(villes):  # default start = 0
    print(f"- {i} - {ville}")
```

```python
for i, ville in enumerate(villes, start=1):
    print(f"- {i} - {ville}")
```

```python
longueurs = []
for ville in villes:
    longueurs.append(len(ville))
print(longueurs)
total_letters = sum(longueurs)
total_letters
```

```python
total_letters = sum(len(ville) for ville in villes)
total_letters
```

```python
# generator (lazy)
g = (len(ville) for ville in villes)
g
```

```python
next(g)
```
